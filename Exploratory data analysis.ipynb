```python
import os
import pandas as pd
import numpy as np
import csv
```

Reading the data of the world happiness report from the files for the years 2015, 2016 and 2017.

```python
data_path = "./../world-happiness-report/"

df_2016 = pd.read_csv(f"{data_path}2016.csv")
df_2017 = pd.read_csv(f"{data_path}2017.csv")
df_2018 = pd.read_csv(f"{data_path}2018.csv")
df_2016.head(5)
```

## Creating combined dataset for all 3 years

    From these csv files, we need to create one output file 'world-happiness-report.csv' which has below columns:
        - country
        - region
        - happiness_score
        - year

### Preprocessing 2016 dataset
Since all these files has different column names, We will convert them into 1 single format.

```python
pd.read_csv('2016.csv', nrows=1).columns
```

```python
op_2016 = df_2016[['Country', 'Region','Happiness Rank', 'Happiness Score','Economy (GDP per Capita)']]
op_2016.insert(5, 'year', 2016)
op_2016 = op_2016.rename(index=str,
                           columns={'Country':'country', 
                                    'Region':'region', 
                                    'Happiness Rank':'Happiness_Rank',
                                    'Happiness Score':'happiness_score',
                                   'Economy (GDP per Capita)':'GDP_Per_Capita'})
op_2016.head(5)
```

### Preprocessing 2017 dataset


```python
pd.read_csv('2017.csv', nrows=1).columns
```

```python
op_2017 = df_2017[['Country', 'Happiness.Rank','Happiness.Score','Economy..GDP.per.Capita.']]
op_2017.insert(4, 'year', 2017)
op_2017 = op_2017.rename(index=str,
                           columns={'Country':'country', 
                                    'Happiness.Rank' :'Happiness_Rank',
                                    'Happiness.Score':'happiness_score',
                                    'Economy..GDP.per.Capita.':'GDP_Per_Capita'
                                   })
op_2017.head(5)
```

### Preprocessing 2018 dataset

```python
pd.read_csv('2018.csv', nrows=1).columns
```

```python
op_2018 = df_2018[['Rank','Country', 'Score','GDP_Per_Capita']]
op_2018.insert(4, 'year', 2018)
op_2018 = op_2018.rename(index=str,
                           columns={'Country':'country', 
                                    'Rank' : 'Happiness_Rank',
                                    'Score':'happiness_score',
                                   'GDP_Per_Capita':'GDP_Per_Capita'})
op_2018.head(5)
```

### Join 2016 and 2018 datasets to get regions coloumn

```python
op_2018_extended = op_2018.merge(op_2016, on='country', how='left', suffixes=['','2016'])
op_2018_extended.head(5)
```

```python
op_2018 = op_2018_extended[['country', 'region','Happiness_Rank','happiness_score','GDP_Per_Capita', 'year']]
op_2018.head(5)
```

### Joining all 3 files to 1 single file

```python
output = pd.concat([op_2016, op_2017, op_2018], ignore_index=True,sort=False)
print(f"Total number of records is {len(output)}")
output.head(5)
```

### Exporting to JSON file


```python
result_file = f"./../world-happiness-report/world-happiness-report.json"
output.to_json(result_file, orient='records', lines=True)
```
